```python
import numpy as np
from cv2 import imread
from google.colab.patches import cv2_imshow
```

```python
img=imread('/content/onerice.png')
cv2_imshow(img)
```

```python
def howIs(img):
  print("tamaño  : ",img.shape)
  print("Maximo  : ",np.max(img))
  print("Minimo  : ",np.min(img))
```

```python
howIs(img)
```

```python
x = img[:,:,1]
howIs(x)
cv2_imshow(x)
```

```python
def segmenta(X,t):
  (M,N) = X.shape
  Y = np.zeros((M,N))
  for i in range(M):
    for j in range(N):
      if (X[i,j]>t):
        Y[i,j] = 255
  return Y
```

```python
prueba = segmenta(x,170)
cv2_imshow(prueba)
```

```python
def area(X,t):
  (M,N) = X.shape
  area = 0
  for i in range(M):
    for j in range(N):
      if (X[i,j]>t):
        area = area + 1
  return area
```

```python
area(x,170)
```

```python
def promedio_gris(X,t):
  (M,N) = X.shape
  Y = np.zeros((M,N))
  contador = 0; suma = 0; promedio = 0
  for i in range(M):
    for j in range(N):
      if (X[i,j]>t):
        suma = suma + X[i,j]
        contador = contador + 1
  
  promedio = suma / contador
  return promedio
```

```python
promedio_gris(x,170)
```
